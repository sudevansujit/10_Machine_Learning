# Pandas.DataFrame.drop_duplicates()

```python
import pandas as pd
import numpy as np
```

```python
airbnb = pd.read_csv('ny_airbnb_data/AB_NYC_2019.csv')
airbnb.head()
```

```python
# Finding Boolean of duplicate rows
airbnb["is_duplicate"]= airbnb.duplicated()
 
airbnb.head()
```

```python
airbnb[airbnb["is_duplicate"] == True]
```

### Looks like there are no duplicate rows in this dataset

```python
#2nd Method

duplicate_bool = airbnb.duplicated(  keep='first')
duplicate = airbnb.loc[duplicate_bool == True]
duplicate
```

## Deleting duplicate Rows

### Pandas has a default function to remove duplicates, drop_duplicates().


```python
#Letâ€™s set the first row equal to the second row and then remove any duplicates.

airbnb.loc[1] = airbnb.loc[0]
airbnb.head()
```

```python
airbnb.drop_duplicates(keep = 'first', inplace = True)
airbnb.head()
```
