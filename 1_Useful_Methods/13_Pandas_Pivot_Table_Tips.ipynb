### Pivot the data

```python
pd.pivot_table(data, index = 'Name')
```

### Multiple indexes



```python
pd.pivot_table(data, index = ['Name', 'Manager'])
```

```python
pd.pivot_table(data, index=["Manager","Rep"])
```

### Explicitly defining the columns we care about using the values field.

```python
pd.pivot_table(data, index=["Manager","Rep"], values = ['Price', 'Quantity'])
```

### Changing aggregate from mean to sum

```python
pd.pivot_table(data, index=["Manager","Rep"], values = ['Price', 'Quantity'], aggfunc = np.sum)
```

```python
pd.pivot_table(data, 
               index=["Manager","Rep"], 
               values = ['Price', 'Quantity'], 
               aggfunc = [np.sum, np.mean, len])
```

```python
# Renember columns are optional Agg func are applied to values
pd.pivot_table(data,
               index=["Manager","Rep"],
               values=["Price"],
               columns=["Product"],
               aggfunc=[np.sum])
```

```python
# Replacing NaN
pd.pivot_table(data,
               index=["Manager","Rep"],
               values=["Price"],
               columns=["Product"],
               aggfunc=[np.sum],
               fill_value=0)
```

```python
pd.pivot_table(data,
               index=["Manager","Rep","Product"],
               values=["Price","Quantity"],
               aggfunc=[np.sum],
               fill_value=0)
```

```python
# To see totals give margins = True
pd.pivot_table(data,
               index = ["Manager","Rep","Product"],
               values = ["Price","Quantity"],
               aggfunc = [np.sum,np.mean],
               fill_value = 0,
               margins = True)
```

### Query Pivot table

```python
table = pd.pivot_table(data,index=["Manager","Status"],columns=["Product"],values=["Quantity","Price"],
               aggfunc={"Quantity":len,"Price":[np.sum,np.mean]},fill_value=0)
table
```

```python
table.query('Manager == ["Debra Henley"]')
```

```python
table.query('Status == ["pending","won"]')
```
