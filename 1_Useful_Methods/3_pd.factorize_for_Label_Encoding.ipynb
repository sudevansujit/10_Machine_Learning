```python
import pandas as pd
import numpy as np
```

### Say we want to label encode the neighbourhood column in the data.


```python
pd.factorize(airbnb['neighbourhood'])[0]
```

```python
pd.factorize(airbnb['neighbourhood'])[1]
```

        The first element of the tuple is an array that contains label-encoded values of the Series. The second element of the tuple is an index/list where each element’s index corresponds to a label in the first element. For example, this first label-encoded value is 0. This means that its reverse-encoded value is the 0th index of the second tuple element, which is Kensington (you can scroll up to confirm this)

```python
airbnb['neighbourhood'] = pd.factorize(airbnb['neighbourhood'])[0].reshape(-1, 1)
airbnb.tail()
```
