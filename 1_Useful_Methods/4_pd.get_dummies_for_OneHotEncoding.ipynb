# Pandas Get_Dummies

```python
import pandas as pd
import numpy as np
```

```python
airbnb = pd.read_csv('ny_airbnb_data/AB_NYC_2019.csv')
airbnb.head()
```

### Say we wanted to One-Hot Encode the same room_type  column from the dataset.

### Method 1 - Direct Method

```python
#For Specific Columns
airbnb1 = pd.get_dummies(airbnb, prefix = ['Room'], columns = ['room_type'])  #Very Important keep inside[Brackets]
airbnb1.head()
```

```python
# dropping first column
#For Specific Columns
airbnb2 = pd.get_dummies(airbnb, prefix = ['Room'], columns = ['room_type'], drop_first= True)  #Very Important keep inside[Brackets]
airbnb2.head()
```

### It even alphabetically arranges the columns for us

### Method 2 Pandas Concatenation - Keeping original column intact

```python
pd.get_dummies(airbnb['room_type'])[:10]
```

```python
room_dummy = pd.get_dummies(airbnb['room_type'], drop_first= True)
room_dummy.head()
```

```python
#Concatenating dummy variable to original data

new_data = pd.concat([airbnb, room_dummy], axis = 1)
new_data.head()
```
