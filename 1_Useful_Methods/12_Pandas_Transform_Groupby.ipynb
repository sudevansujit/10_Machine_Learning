### Transform

`Transform is an operation used in conjunction with groupby`

```python
data.groupby('order')["ext price"].sum()
```

```python
data.groupby('order')['ext price'].transform('sum')
```

```python
# combining order and total order using transform
data['total order'] = data.groupby('order')['ext price'].transform('sum')
data
```

```python
# Percentage of order
data["Percent_of_Order"] = data["ext price"] / data["total order"]
data
```
