# Pandas Sidetable

```python
import pandas as pd
import numpy as np
```

```python
bikes = pd.read_csv('day.csv')
bikes.head()
```

```python
import sidetable
```

```python
bikes['weekday'].value_counts()
```

```python
bikes['weekday'].value_counts(normalize=True)
```

```python
bikes.stb.freq(['weekday'])
```

```python
for i in list(bikes.columns):
    print(bikes.stb.freq([i]))
```

### We can set a threshold on cumulative Percentage

```python
bikes.stb.freq(['weekday'], thresh= 0.7, other_label='other days')
```

### Pass Multiple columns to see more Specific Distribution

```python
bikes.stb.freq(['season','yr'])
```

```python
bikes.stb.freq(['season','yr'], thresh=.5)
```

```python
bikes[bikes.season == 1].stb.freq(['weekday'])
```

```python
bikes[bikes.season == 4].stb.freq(['weekday'])
```

```python
bikes.stb.freq(['weathersit'], value='windspeed', style = True)
```
